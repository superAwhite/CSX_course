## 有發現到縣市填寫不完全/亂填，建議註冊會員時用表單方式

```python
dict_region = {"北部":['臺北','台北','新北','基隆','宜蘭','桃園','新竹'],
               "中部":['苗栗','臺中','台中','彰化','南投','雲林'],
               "南部":['嘉義','臺南','台南','高雄','屏東','澎湖'],
               "東部":['花蓮','台東','臺東']
              }
```

```python
m_state = members[members['LocationState'] == members['LocationState']]  #過濾nan
m_state
```

```python
def get_keys(d, seri_state):
    
    result = list()
    for state in seri_state:
        locate_flag = 0
        for key,value in d.items():
            for a in value:            
                if a in state:
                    result.append(key)
                    locate_flag = 1
                    break
            if locate_flag == 1:
                break
        if locate_flag == 0:
            result.append('not_define')
    return result
    
get_keys(dict_region, ['桃園'])
```

```python
m_state['Area'] = get_keys(dict_region, m_state['LocationState'])
```

```python
m_state =  m_state[m_state['Area'] != 'not_define']

```

```python
import datetime
today = datetime.date.today()
today.year

m_state['new_Birthday'] = pd.to_datetime(m_state['Birthday'].astype(str), format='%Y-%m-%d', errors='ignore')#濾掉生日亂填的

```

```python
m_state = m_state[m_state['new_Birthday'] != 'nan']
len(m_state)
```

```python
a = list(m_state['new_Birthday'])

m_state['age'] = [today.year-datetime.datetime.strptime(a[i],'%Y-%m-%d').year
                  for i in range(len(a))]    
m_state = m_state[(m_state['age']>0) & (m_state['age']<150)]
```

```python
m_numofarea = pd.DataFrame(m_state.groupby(by = 'Area')['Frequency','Monetary'].sum())
m_numofarea
```
